```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./image_chat.png")
```

```python
def cut_image(image, width, height):
    array = image[:width, :height]
    arr = []
    for i in range(len(array)):
        for j in range(len(array[i])):
            arr.append(np.mean(array[i][j]))
    for i in range(len(arr)):
        arr[i] = round(arr[i])
    return np.reshape(arr, (8, 8))
```

```python
array = cut_image(cv2.imread("./image_chat.png"), 8, 8)
print(array)
plt.imshow(array)
```

```python
def gray(r, g, b):
    return 0.2126 * r + 0.7152 * g + 0.0722 * b

def convert_to_grayscale(image):
    res = image
    for i in range(len(image)):
        for j in range(len(image[i])):
            res[i][j] = [gray(image[i][j][0], image[i][j][1], image[i][j][2])]
    return res

plt.imshow(convert_to_grayscale(image))
```
